# Visualizing Statistical relationships
  <ol><li>Scatter plot
      <li>hue plot

## 1. Scatterplot using seaborn-relplot()

```python
sns.relplot(x='age',y='length_of_service',data=df)
```

In above plot we can see that age increases length of services also increases

## 2. Hue plot-Hue can be any variable 

```python
sns.relplot(x='age',y='length_of_service',hue="gender",data=df)
```

In above plot we can see the no of males and females with the help of 'hue' parameter

# 2. Plotting Categorical Data
  <ol><li>jitter-catplot(x,y,data,hue,kind,jitter,col,split,aspect)
      <li>Hue
      <li>Boxplot
      <li>Voilin plot
      <li>Pointplot
  

```python
sns.catplot(x='education',y='avg_training_score',data=df)
```

```python
#As the plot is scatter we can also disable it 
sns.catplot(x='education',y='avg_training_score',jitter=False,data=df)
```

<p>We can see that for masters & above  and bachelors degree people share same training score

```python
#As the plot is scatter we can also disable it 
sns.catplot(x='education',y='avg_training_score',hue='is_promoted',jitter=True,data=df)
```

<p>And now by passing hue we can see the no of promoted and not promoted employee also

```python
#sns.catplot(x='education',y='avg_training_score',kind="swarm",hue='gender',data=df)
```

kind="Swarm" prevents overlapping and produces clean scatter plot

```python
sns.boxplot(x='education',y='avg_training_score',hue='is_promoted',data=df)
```

<p>Above plot shows the median and all measures of people promoted with respect to there education

<h1>Violin plots</h1>
<p>The violin plots combine the boxplot and kernel density estimation procedure to provide richer description of the distribution of values. The quartile values are displayed inside the violin.

```python
ax=sns.catplot(x='education',y='avg_training_score',hue='is_promoted',kind='violin',data=df)
ax.set_xticklabels(rotation=30)
```

```python
ax=sns.catplot(x='education',y='avg_training_score',hue='is_promoted',kind='violin',split=True,data=df)
ax.set_xticklabels(rotation=30)
```

<p>Split works only when hue has only two levels

```python
ax=sns.catplot(x='department',y='avg_training_score',data=df,kind="bar",hue='is_promoted')
ax.set_xticklabels(rotation=30)
```

<h1>Point Plot</h1>
<p>This plot points out the estimate value and confidence interval

```python
ax=sns.catplot(x='department',y='avg_training_score',data=df,kind="point",hue="is_promoted")
ax.set_xticklabels(rotation=90)
```

# 3. Plotting Bivariate Distributions-jointplot()
 <ol><li>Hexplot
      <li>KDE Plot
      <li>Boxen Plot
      <li>Ridge Plot
    
 

```python
sns.jointplot(x='avg_training_score',y='age',data=df)
```

Above plot shows the data distribution of two variables

<h1>Hex Plot</h1>
<p>Hexplot is a bivariate analog of histogram as it shows the number of observations that falls within hexagonal bins. This is a plot which works with large dataset very easily. To draw a hexplot, we’ll set kind attribute to hex.</p>

```python
sns.jointplot(x='age',y='avg_training_score',kind="hex",data=df)
```

<h1>KDE Plot</h1>


```python
sns.jointplot(x='age',y='avg_training_score',kind="kde",data=df)
```

# Heat Map
